### This notebook will produce two simple plots of the relationship between redshift and age of the universe (t)

```python
#Importing libraries necessary to make plots, manipulate arrays, etc.
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# creating arrays for values of redshift (z) and age of the universe (t)
#These values are taken from the 
z = np.array([0.1, 0.2, 0.4, 0.6, 0.8, 1.6, 3.2, 6.4, 10])
t = np.array([12.15, 11.02, 9.17, 7.74, 6.62, 3.95, 1.95, 0.83, 0.45])
```

```python
# testing array creation
print z
print t
```

```python
# Plotting redshift v. age
plt.plot(t, z)
plt.xlim(0, 12.2)
plt.ylabel('Redshift')
plt.xlabel('Age of the Universe $(Gyr)$')
plt.savefig('Redshift v. Age.png')
```

```python
# Plotting age v. redshift
plt.plot(z, t, 'r')
plt.ylim(0, 12.2)
plt.xlabel('Redshift')
plt.ylabel('Age of the Universe $(Gyr)$')
plt.savefig('Age v. Redshift.png')
```
